```python
#importing all the libraries that we need
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd 
```

```python
csvFile = open("PracticalSalinityUnits_Dec_2011.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
salinity_2011 = list(csvReader)
print (salinity_2011)
```

```python
width= (len(salinity_2011[0]))
height= (len(salinity_2011))
print (width)
print (height)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
salinity_map_2011 = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(salinity_2011):
    for c, value in enumerate(row):
        if float(value) < 99999:
            salinity_map_2011.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 30: #Blue, less than or equal to 0
               grid[r, c] = [2, 59, 247]
        elif float(value) <= 31: #Teal, less than or equal to 2
               grid[r, c] = [1, 242, 247]
        elif float(value) <= 32: #Purple, less than or equal to 5
               grid[r, c] = [121, 16, 150]
        elif float(value) <= 33: #Dark green, less than or equal to 10
               grid[r, c] = [1, 58, 16]
        elif float(value) <= 34: #Light green, less than or equal to 15
               grid[r, c] = [49, 224, 96]
        elif float(value) <= 35: #Yellow, less than or equal to 20
               grid[r, c] = [238, 242, 4]
        elif float(value) <= 36: #Orange, less than or equal to 25
               grid[r, c] = [239, 138, 7]
        elif float(value) <= 37: #Light red, less than or equal to 30
               grid[r, c] = [244, 73, 73]
        elif float(value) <= 38: #Dark red, less than or equal to 35
               grid[r, c] = [173, 0, 0]
        else:
               grid[r, c] = [0, 0, 0]
        
```

```python
print(min(salinity_map_2011))
print(max(salinity_map_2011))
```

```python
plt.title("Salinity of the Ocean December 2011")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Land_temperature.png",dpi = 1000)
plt.show()
```

```python
df= pd.read_csv("Salinity_info_38N_68W.csv")
```

```python
df.head(2)
```

```python
depth= df["Depth "]
temp= df ["Temperature (Celsius)"]
salinity= df["Salinity (PSU)"]
density= df["Density (kg/m3)"]
```

```python
print (depth.head())
print (temp.head())
print (salinity.head())
print (density.head())
```

```python
# the plt.clf function prevents the graph from abnormalities, by clearing the figure
plt.clf()

#code for the temperature line and the location of its legend
plt.plot(depth, temp, color="green")
plt.legend(loc=2)

#in order for the other two x and y labels to show up on te graph PUT IT BEFORE THE plt.twinx FUNCTION!
plt.ylabel("temperature(celsius)")
plt.xlabel("depth")

#plt.twinx function used to combine the graph for temp and salinity

plt.twinx()

#code for the salinty line and the location of its legend
plt.plot(depth, salinity, color="blue")
plt.legend()

#plot title
plt.title("Salinity and Temperature")

# changing these numbers changes the range on the graph
plt.xlim(0, 4500)
plt.ylim(0,40)

# the label for the 3rd side of the graph
plt.ylabel("salinity(psu)")

#where do you want the tick marks to be
plt.xticks(range(0, 4500,500))
plt.show()
```
