```python
#this function is used to create marker sizes 
def sizeFunc(column, numSize):    
        empty_list = []
        interval = (column.max()- column.min())/numSize
        num_temp = column.min()
        while len(empty_list) < numSize:
            num_temp += interval
            empty_list.append(num_temp)
        return empty_list
sizeFunc(quake_df["EQ_PRIMARY"], 6)
mage = quake_df["EQ_PRIMARY"]
sizeFunc(vol_df["VEI"], 6)
mag = vol_df["VEI"]


# this assignes the size of the marker from the range that we made
def Func(column):
    empty_list = sizeFunc(vol_df["VEI"], 6) 
    sizes = []                               
    for item in column:                      
        if item <= empty_list[0]:
            sizes.append(7)
        elif item <= empty_list[1]:
            sizes.append(30)
        elif item <= empty_list[2]:
             sizes.append(70)
        elif item <= empty_list[3]:
            sizes.append(100)
        elif item <= empty_list[4]:
            sizes.append(200)
        elif item <= empty_list[5]:
            sizes.append(300)
    return sizes
s = Func(mag)

# this code is where everything comes together in the same map 
image = plt.imread("earth.jpg")
fig, ax = plt.subplots()
ax.imshow(image, extent = [-180,180,-90,90])
plt.scatter(earth_long, earth_lat, color = colorList, s = s, marker = "*")
plt.scatter(fire_long, fire_lat, color = colorList, s = s, marker = "^" )
plt.xlim(-180, 180)
plt.ylabel("Latitude")
plt.xlabel("Longitude")
plt.title("Volcanoes and Earthquakes since 2000")
plt.savefig("volcanoes_and_Earthquakes.png",dpi = 1000)
```
